DMC and BUI are highly correlated. Therefore DMC can be removed in this data

```python
df=df.drop('DMC', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Region']=le.fit_transform(df['Region'])
df['classes']=le.fit_transform(df['classes'])
```

```python
Q1 = df['Rain '].quantile(0.15)
Q3 = df['Rain '].quantile(0.95)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
```

```python
df['Rain '][df['Rain ']<lower_limit]=lower_limit
df['Rain '][df['Rain ']>upper_limit]=upper_limit
```

```python
fig, axs = plt.subplots(ncols=5, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in X.items():
    sns.boxplot(y=k, data=X, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
X_train.shape
```

```python
y_train.shape
```

implemeneting simple random forest algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
paramgrid = {
    'max_depth': list(range(2, 10, 2)),
    'criterion': ["gini", "entropy"]
}
grid_search=GridSearchCV(DecisionTreeClassifier(),paramgrid)
```

```python
grid_search.fit(X_train,y_train)
```

checking the best scores and parameters from our hyperparameter tuning
